```python
import snscrape.modules.telegram as sntwitter
import json
import os
from datetime import date
import itertools
import pandas as pd
```

```python
telegram_messages_list = []
```

```python
#Import List of Keywords from Txt File
# opening the file in read mode
my_file = open("telegram_channels.txt", "r")

# reading the file
keyword_data = my_file.read()

# replacing end splitting the text 
# when newline ('\n') is seen.
telegram_channel_list = keyword_data.split("\n")
print(telegram_channel_list)
```

```python
for telegram_channel in telegram_channel_list:
    tg_df = pd.DataFrame(itertools.islice(sntwitter.TelegramChannelScraper(f'{telegram_channel}').get_items(), 20))
    telegram_messages_list.append(tg_df)
```

```python
telegram_messages_df = pd.concat(telegram_messages_list)
```

```python
telegram_messages_df
```

```python
telegram_messages_df = telegram_messages_df[['date', 'content', 'url']]
telegram_messages_df.reset_index(inplace=True, drop=True)
telegram_messages_df
```

```python
telegram_messages_df = telegram_messages_df.dropna()
telegram_messages_df.reset_index(inplace=True, drop=True)
telegram_messages_df
```

```python
#Import List of Keywords from Txt File
# opening the file in read mode
my_file = open("kucoin_list.txt", "r")

# reading the file
kucoin_data = my_file.read()

# replacing end splitting the text 
# when newline ('\n') is seen.
kucoin_list = kucoin_data.split("\n")
```

```python
#Filter tweets to only include Kucoin coins
telegram_messages_kucoin_df = telegram_messages_df[telegram_messages_df['content'].str.contains('|'.join(kucoin_list),regex=True)]
telegram_messages_kucoin_df.reset_index(inplace=True, drop=True)
telegram_messages_kucoin_df
```

```python
telegram_messages_kucoin_df = telegram_messages_df.set_index('date').sort_index()
telegram_messages_kucoin_df
```

```python
my_list = []
for row in telegram_messages_df['content']:
    tokens = row.split(" ")
    for token in tokens:
        if token in kucoin_list:
            my_list.append(token)
        else:
            pass
```

```python
from collections import Counter
counts = Counter(my_list)
best_coins = counts.most_common(50)
best_coins
```

```python
def get_crypto(crypto):
    coin = telegram_messages_kucoin_df[telegram_messages_kucoin_df['content'].str.contains(crypto)]
    return coin
```
