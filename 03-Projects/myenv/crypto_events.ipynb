```python
import twint
import nest_asyncio
import pandas as pd
from pathlib import Path
nest_asyncio.apply()
from datetime import date
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
```

```python
#Import List of Keywords from Txt File
my_file = open("remove_keywords.txt", "r")
keyword_data = my_file.read()
remove_list = keyword_data.split("\n")
print(remove_list)
```

```python
#Import List of Keywords from Txt File
# opening the file in read mode
my_file = open("kucoin_list.txt", "r")

# reading the file
kucoin_data = my_file.read()

# replacing end splitting the text 
# when newline ('\n') is seen.
kucoin_list = kucoin_data.split("\n")
```

```python
#Get yesterday's date
yesterday_datetime = datetime.now() - timedelta(days=1)

yesterday = str(yesterday_datetime.strftime('%Y-%m-%d'))

print(yesterday)
```

```python
def crypto_tweets(username):
    c = twint.Config()
    c.Username = username
    c.Since = yesterday
    c.Pandas = True
    twint.run.Search(c)
```

```python
crypto_tweets("cmcal_bot")
```

```python
todays_cmc_tweets_df = twint.storage.panda.Tweets_df
```

```python
todays_cmc_tweets_df = todays_cmc_tweets_df[['date','tweet','link']]
todays_cmc_tweets_df.reset_index(inplace=True, drop=True)
todays_cmc_tweets_df
```

```python
#Remove Irrelevant events using the remove keyword list
todays_cmc_tweets_df = todays_cmc_tweets_df[~todays_cmc_tweets_df['tweet'].str.contains('|'.join(remove_list),regex=True)]
todays_cmc_tweets_df.reset_index(inplace=True, drop=True)
todays_cmc_tweets_df
```

```python
#Filter tweets to only include Kucoin coins
todays_cmc_tweets_df = todays_cmc_tweets_df[todays_cmc_tweets_df['tweet'].str.contains('|\$'.join(kucoin_list),regex=True)]
todays_cmc_tweets_df.reset_index(inplace=True, drop=True)
todays_cmc_tweets_df.to_csv('todays_events.csv')
todays_cmc_tweets_df
```

```python
my_list = []
for row in todays_cmc_tweets_df["tweet"]:
    tokens = row.split(" ")
    for token in tokens:
        if token.startswith("$"):
            my_list.append(token)
        else:
            pass
```

```python
from collections import Counter
counts = Counter(my_list)
best_coins = counts.most_common(20)
best_coins_df = pd.DataFrame(best_coins)
best_coins_df.set_index(0, inplace=True)
plt.figure(figsize=(17,4))
plt.bar(x=best_coins_df.index, height=best_coins_df[1], color='tan')




```

```python
def get_crypto(crypto):
    coin = todays_cmc_tweets_df[todays_cmc_tweets_df['tweet'].str.contains(crypto)]
    coin.reset_index(inplace=True, drop=True)
    return coin
```
